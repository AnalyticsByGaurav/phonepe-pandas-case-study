#5.1: Plot the total transactions and amount over time for a selected state
    Create a line plot showing the total number of transactions and the total transaction
    amount over time (years and quarters) for a selected state.

```python
def plot_state_trend(state_name):
    
    state_df = (
        df_State_Txn_Users[["State", "Year", "Quarter", "Transactions", "Amount (INR)"]]
        [df_State_Txn_Users["State"] == state_name]
        .groupby(["Year", "Quarter"], dropna=False)
        .agg(Total_Transactions=("Transactions", "sum"),Total_Amount=("Amount (INR)", "sum"))
        .reset_index()
    )
    
    state_df["Year_Quarter"] = (state_df["Year"].astype(str) + "-Q" + state_df["Quarter"].astype(str))
    #print(state_df)
    plt.figure(figsize=(12, 6))
    plt.plot(state_df["Year_Quarter"], state_df["Total_Transactions"], label="Total Transactions")
    plt.plot(state_df["Year_Quarter"], state_df["Total_Amount"], label="Total Amount")
    plt.xlabel("Year - Quarter")
    plt.ylabel("Value")
    plt.title(f"{state_name}: Transactions & Amount Over Time")
    plt.xticks(rotation=45)
    plt.legend()    
    plt.show()
```

```python
plot_state_trend("Uttar Pradesh")
```

```python
df_State_Txn_Users["State"].unique()
```

```python
plot_state_trend("West Bengal")
```

#5.2 Create a pie chart showing the distribution of different transaction types for a selected state and quarter.

```python
def plot_txn_type_pie(state_name, year, quarter):
    
    pie_df = (
        df_State_TxnSplit[["State", "Year", "Quarter", "Transaction Type", "Transactions"]]
        [(df_State_TxnSplit["State"] == state_name) &
        (df_State_TxnSplit["Year"] == year)         &
        (df_State_TxnSplit["Quarter"] == quarter)]
        .groupby("Transaction Type", dropna=False)
        .agg(Total_Transactions=("Transactions", "sum"))
        .reset_index()
    )
    plt.figure(figsize=(6, 6))
    plt.pie(pie_df["Total_Transactions"],labels=pie_df["Transaction Type"],autopct="%1.1f%%")
    plt.title(f"{state_name} – Transaction Type Distribution ({year} Q{quarter})")
    plt.tight_layout()
    plt.show()
```

```python
plot_txn_type_pie('Uttar Pradesh',2020,4)
```

#5.3 Create a bar plot showing the population density of districts in a selected state.

```python
def plot_population_density(state_name):

    plot_df = (
        district_merged[
            ["State", "District", "Population_Density"]
        ]
        [district_merged["State"] == state_name]
        .sort_values("Population_Density", ascending=False)
    )

    plt.figure(figsize=(14, 7))
    plt.bar(plot_df["District"], plot_df["Population_Density"])
    plt.xlabel("District")
    plt.ylabel("Population Density (People per sq km)")
    plt.title(f"District-wise Population Density in {state_name}")
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()
```
